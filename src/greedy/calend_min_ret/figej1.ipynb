```python
from cmath import atan
from glob import glob
from matplotlib.patches import Circle
from matplotlib.patches import Rectangle
from matplotlib.patches import Path
from matplotlib.patches import PathPatch
from matplotlib.patches import Ellipse 
import matplotlib.pyplot as plt
from IPython.display import display
import ipywidgets as widgets
from IPython.display import clear_output
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import random 
%matplotlib nbagg

class figReg1(): 
    def vertl(self,x,y1,y2): 
        l = PathPatch(Path([(x,y1),(x,y2)]),linestyle = '--')
        self.ax.add_patch(l)
    def rect(self,x,y,w,h): 
        r1 = Rectangle((x,y),width = w,height = h,facecolor = 'white',edgecolor = 'black')
        self.ax.add_patch(r1)
    def poner_ejex(self): 
        l = PathPatch(Path([(0,-1),(20,-1)]))
        self.ax.add_patch(l)
    def poner_numeje(self): 
        self.ax.text(0,-2,"$0$",va='center',ha='center')
        self.ax.text(10,-2,"$10$",va='center',ha='center')
        self.ax.text(20,-2,"$20$",va='center',ha='center')
    def anotar_deadline(self): 
        self.ax.text(10.5,-.5,"$d_1$")
        self.ax.text(20.5,-.5,"$d_2$")
    def poner_intervalos(self):
        self.rect(0,0,1,1)
        self.rect(0,3,10,1)
        self.ax.text(2,0.5,"$i_1$",va='center',ha='center')
        self.ax.text(11,3.5,"$i_2$",va='center',ha='center')
        self.vertl(20,-1,5)
        self.vertl(10,-1,5)
        self.poner_ejex()
        self.poner_numeje()
        self.anotar_deadline()
    def poner_calend1(self): 
        def poner_intervs():
            r1 = Rectangle((0,-10),width = 1,height = 1,facecolor = 'white',edgecolor = 'black')
            self.ax.add_patch(r1)
            r2 = Rectangle((1,-10),width = 10,height = 1,facecolor = 'white',edgecolor = 'black')
            self.ax.add_patch(r2)
            self.ax.add_patch(PathPatch(Path([(0,-11),(20,-11)])))
            self.ax.add_patch(PathPatch(Path([(10,-11),(10,-5)]),linestyle = '--'))
            self.ax.add_patch(PathPatch(Path([(20,-11),(20,-5)]),linestyle = '--'))
            self.ax.add_patch(PathPatch(Path([(10,-8),(11,-8)]),color = 'red'))
            self.ax.text(22,-7,"(a)")
        poner_intervs() 
    def poner_calend2(self): 
        def poner_intervs():
            r2 = Rectangle((30,-10),width = 10,height = 1,facecolor = 'white',edgecolor = 'black')
            self.ax.add_patch(r2)
            r1 = Rectangle((40,-10),width = 1,height = 1,facecolor = 'white',edgecolor = 'black')
            self.ax.add_patch(r1)
            self.ax.add_patch(PathPatch(Path([(30,-11),(50,-11)])))
            self.ax.add_patch(PathPatch(Path([(40,-11),(40,-5)]),linestyle = '--'))
            self.ax.add_patch(PathPatch(Path([(50,-11),(50,-5)]),linestyle = '--'))
            self.ax.text(53,-7,"(b)")
        poner_intervs() 
    def __init__(self): 
        self.fig,self.ax = plt.subplots()
        self.ax.set_aspect('equal', adjustable='box')
        plt.axis("off")
        self.poner_intervalos() 
        self.poner_calend1()
        self.poner_calend2() 
        self.ax.add_patch(PathPatch(Path([(-5,-4),(55,-4)]),color = 'gray'))
        self.ax.relim()
        self.ax.autoscale_view()
f = figReg1() 
```
