<h1> Quicksort </h1> 
<p> 
Quicksort al igual que Mergesort es un algoritmo para ordenar que pertence al paradigma divide y vencerás. Quicksort tiene un cota en el peor de los casos $O(n^2)$ sin embargo, para el caso promedio la cota es $O(n \log(n))$. El algoritmo también tiene la ventaja de ordenar in-place
</p>  

<h2> la rutina de división de Quicksort</h2> 
<p> 
Como ya vimos los algoritmos del paradigma divide y vencerás tienen 3 partes: dividir, conquistar y combinar. 
Como veremos más adelante, cuando se presente el algoritmo completo, la parte importante de quicksort es dividir. 
La rutina para dividir toma, al inicio, un pivote y separa los números que son menores o iguales a tal pivote en la primera parte del arreglo y los números que son mayores o iguales en la parte restante. Para que el tiempo de ejecución sea $O(n\log(n))$ en el caso promedio esta rutina debe tener un tiempo de ejecución lineal. A continuación se muestra el pseudocódigo de la rutina sobre un arreglo seguido de la celda que ilustra su comportamiento. 
</p> 

<div style="background-color:rgba(0, 0, 0, 0.0470588); padding:10px 0;font-family:monospace;">
<b> Subrutina para dividir($A$) </b> <br> 
&nbsp;$r = l(A)-1$ <br> 
&nbsp;$x = A[r]$ <br> 
&nbsp;$i = -1$ <br> 
&nbsp;for $j = 0$ to $r -1$ <br> 
&nbsp;&nbsp;if $A[j] \leq x$ <br> 
&nbsp;&nbsp;&nbsp;$i = i + 1$ <br> 
&nbsp;&nbsp;&nbsp;intercambiar $A[i]$ con $A[j]$ <br> 
&nbsp;intercambiar $A[i+1]$ con $A[r]$ <br> 
</div>
