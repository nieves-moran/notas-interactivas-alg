<h1>El problema</h1> 
<p> 
Supongamos que tenemos una máquina y un conjunto de peticiones $\{1\dots n\}$. Solo podemos usar este recurso desde el tiempo 0 hasta un cierto tiempo $W$. Cada petición corresponde a trabajo que requiere tiempo $w_i$ para realizarse. El objetivo es realizar tantos 
    trabajos como sea posible. 
    </p> 
    <p> 
    El problema anterior puede ser formalizado de la siguiente manera: Tenemos un conjunto $\{w_1,w_2,\dots , w_n\}$
    y un número positivo $W$. Queremos seleccionar un subconjuto $S$ talque $\sum_{i \in S}w_i \leq W$ y que además 
    $\sum_{i\in S}w_i$ sea tan grande como sea posible. Este problema es llamado Subset Sum. 
    </p>
