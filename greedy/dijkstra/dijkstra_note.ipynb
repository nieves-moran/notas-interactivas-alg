<h1> Caminos más cortos en una gráfica </h1> 
Un problema muy basico es determinar el camino mas corto entre dos nodos en una grafica. O mas informacion 
dado un nodo de inicio, cual el camino mas corto de ese nodo a los demoas . 
Tenemos una grafica uqe es G=(V,E) con un nodo s designano. cada arista tiene un peso, que es la distancia o el costo 
tambien llamado. El objetivo es hacer el camino de el nodo a todos los demas de la grafica. 
<h1> Diseñando el Algoritmo</h1>
    </span>
<p> 
El algoritmo fue diseñado por Edsger Dijkstra en 1959. Mantiene un conjunto $S$ de vertices para los
    cuales ha sido determinado ya el camino más corto partiendo de $s$. Esta es la parte explorada 
    de la gráfica. Ahora para cada nodo que está en la parte no explorada $v \in V - S$, determinamos el 
    camino más corto que puede ser construido viajando por la parte explorada a un nodo $u \in S$ seguido 
    de la arista $(u,v)$. Esto es, consideramos la cantidad $d'(v) = min_{e = (u,v):u \in S}d(u) + l_e$. 
    Escogemos el nodo $v \in V - S$ para el cual esta cantidad es minimizada, añadimos $v$ a $S$ y definimos 
    actualizamos el valor de $d(v)$ como $d'(v)$ 
</p>
