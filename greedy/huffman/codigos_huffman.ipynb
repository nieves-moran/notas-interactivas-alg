<h1> Códigos de Huffman </h1> 

## Introducción al problema
<p> 
    El problema es una pregunta básica en el área de la compresión de datos. 
Dado que las computadoras operan sobre secuencias de bits, tenemos que codificar los símbolos del alfabeto a estas secuencias. La forma más simple es usando un número fijo de bits,sin embargo esto es un desperdicio. Para ver por qué toma en cuenta lo siguiente: 
</p> 
<p> 
    Las siguientes son dos codificaciones del alfabeto ${a,b,c,d}$. La primera es una codificación de longitud fija y la segunda de longitud variable. 
</p> 
<p> Codificación 1 </p> 
\begin{equation}
\begin{split}
     \gamma_1(a)  & = 00 \\ 
     \gamma_1(b) & = 01 \\ 
     \gamma_1(c) & = 10 \\ 
     \gamma_1(d) & = 11 
\end{split}
\end{equation}
<p> Codificación 2</p> 
\begin{equation}
\begin{split}
     \gamma_2(a)  & = 1 \\ 
     \gamma_2(b) & = 001 \\ 
     \gamma_2(c) & = 000 \\ 
     \gamma_2(d) & = 01 
\end{split}
\end{equation}
<p> Ahora si $s = aaabcda$ la codificación de $s$ usando $\gamma_1$ es $\gamma_1(s) = 00000001101100$. En cambio al utilizar $\gamma_2$ tenemos $\gamma_2(s) = 111001000011$. De modo que la longitud de la segunda codificación es menor y por lo tanto mejor.</p> 
<p>
Que la segunda codificación del ejemplo anterior fuese mejor, no fue una casualidad. Resultó mejor ya que la letra $a$ que tenía más frecuencia en $s$ fue asignada a una secuencia de bits más corta. Veremos que lo más óptimo es asignar secuencias de bits más cortas a letras menos frecuentes. Con esto, reducimos la longitud de la códificación de las cadenas en promedio. La pregunta es: ¿Cómo hacemos esta asignación?. 
</p>




## Códigos prefijos
<p> Considera el siguiente código: </p> 
\begin{equation}
\begin{split}
     \gamma(a)  & = 10 \\ 
     \gamma(b) & = 1 \\ 
     \gamma(c) & = 11 \\  
\end{split}
\end{equation}
<p> 
    Ahora supón que es recibida la cadena $s = 1011$. Tal cadena tiene dos posibles decodificaciones: $ac$ y $abb$. Esta ambigüedad es debida a que existe un par de letras donde la cadena de bits que codifica a una es prefijo de la cadena de bits que codifica a otra. Para eliminar está ambigüedad es necesario que el código sea un código prefijo, definido a continuación. 
</p> 
<p> 
    Un <b> código prefijo </b> para un conjunto $S$ de letras es una función $\gamma$ que mapea cada letra $x \in S$ a una secuencia de ceros y unos de manera que para $x, y \in S$ distintas $\gamma(x)$ no es un prefijo de la secuencia $\gamma(y)$     
</p> 

## Códigos prefijos óptimos
<p> 
    La frecuencia de la letra $x$ denotada como $f_x$ es la fracción de letras en el texto que son iguales a $x$. En otras palabras, $n$ es la longitud del texto, $nf_x$ es el número de letras en el texto que son iguales a $x$ 
</p> 
<p> 
    Ahora, Si utilizamos un código $\gamma$, la longitud de la códificación es la siguiente:
   
$$ \sum_{x \in S} n f_x\left|(\gamma(x))\right| = n\sum_{x\in S}f_x\left|\gamma(x)\right|$$
    Y entonces, eliminando el coeficiente $n$ obtenemos el promedio de bits por letra. Denotado por $ABL(\gamma)$ 
    $$ ABL(\gamma) = \sum_{x\in S}f_x\left|\gamma(x)\right|$$
</p> 
<p>
    Luego, podemos poner en otros términos el objetivo inicial. Dado un alfabeto y un conjunto de frecuencias para las letras, queremos el código prefijo más eficiente posible. Esto es, que minimize el promedio de bits por letra ($ABL(\gamma)$). A tal código prefijo le llamamos óptimo. 
</p> 

## Diseñando el algoritmo 

<p> 
    Supón que tenemos un arbol $T$ en el que todo nodo que no es una hoja tiene a lo más 2 hijos. Además, el número de hojas es igual al tamaño del alfabeto y cada hoja la etiquetamos con un símbolo de $S$
</p> 
<p> 
    Este árbol binario $T$ describe un código prefijo de la siguiente manera. Para cada letra $x \in S$ seguimos el camino de la raíz a la hoja etiquetada con la letra $x$. Cada vez que el camino vaya de un nodo a su hijo izquierdo agregamos un 0 a la códificación y cada vez que vaya de un nodo a su hijo derecho,agregamos un 1.  
</p>
