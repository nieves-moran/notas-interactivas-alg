```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.table import table
from matplotlib.transforms import Bbox
from matplotlib.patches import Rectangle
import ipywidgets as widgets 
%matplotlib nbagg 
def create_array_rep(arr,x,y,w,h,ax): 
    rep = [None for i in range(0,len(arr))]
    for i in range(0,len(arr)): 
        r = Rectangle((x, y), w, h,edgecolor = 'black',facecolor = 'white')
        ax.add_patch(r)
        rx, ry = r.get_xy()
        cx = rx + r.get_width()/2.0
        cy = ry + r.get_height()/2.0
        anot = ax.annotate(str(arr[i]), (cx, cy), color='black', weight='bold', fontsize=7, ha='center', va='center')
        rep[i] = (r,anot)
        x = x + w
    return rep

class state: 
    def __init__(self,arr1,arr2,ax): 
        self.k = 0; 
        self.i = 0; 
        self.j = 0; 
        self.arr1 = arr1
        self.arr2 = arr2 
        self.arr1_im = create_array_rep(arr1,1,1,4,3,ax)
        self.arr2_im = create_array_rep(arr2,1,7,4,3,ax)
        self.comb = [0 for i in range(0,len(arr1)+len(arr2))]
        self.comb_im = create_array_rep(self.comb,1,12,4,3,ax)
        for i in range(0,len(self.comb)): 
            self.comb_im[i][1].set(text = "")
        if(self.i <  len(self.arr1)): 
                self.arr1_im[self.i][0].set(facecolor = 'green')
        if(self.j <  len(self.arr2)): 
                self.arr2_im[self.j][0].set(facecolor = 'green')
    def update(self):
        if(self.i == len(self.arr1) and self.j == len(self.arr2) ):
            return
        if(self.i == len(self.arr1)):
            self.comb_im[self.k][1].set(text = str(self.arr2[self.j]))
            self.comb_im[self.k][0].set(facecolor = 'red')
            self.arr2_im[self.j][0].set(facecolor = 'yellow')
            self.j = self.j + 1
            self.k = self.k + 1
            if(self.j <  len(self.arr2)): 
                self.arr2_im[self.j][0].set(facecolor = 'green')
            return 
        if(self.j == len(self.arr2)): 
            self.comb_im[self.k][1].set(text = str(self.arr1[self.i]))
            self.comb_im[self.k][0].set(facecolor = 'red')
            self.arr1_im[self.i][0].set(facecolor = 'yellow')
            self.i = self.i + 1 
            self.k = self.k + 1
            if(self.i <  len(self.arr1)): 
                self.arr1_im[self.i][0].set(facecolor = 'green')
            return
        if(self.arr1[self.i] <= self.arr2[self.j]): 
            self.comb_im[self.k][1].set(text = str(self.arr1[self.i]))
            self.comb_im[self.k][0].set(facecolor = 'red')
            self.arr1_im[self.i][0].set(facecolor = 'yellow')
            self.i = self.i + 1 
            if(self.i <  len(self.arr1)):
                self.arr1_im[self.i][0].set(facecolor = 'green')
        else: 
            self.comb_im[self.k][1].set(text = str(self.arr2[self.j]))
            self.comb_im[self.k][0].set(facecolor = 'red')
            self.arr2_im[self.j][0].set(facecolor = 'yellow')
            self.j = self.j + 1
            if(self.j <  len(self.arr2)): 
                self.arr2_im[self.j][0].set(facecolor = 'green')
        
        self.k = self.k + 1
class cell3: 
    def __init__(self): 
        self.ax = None
        self.maxim_x = None 
        self.maxim_y = None 
        self.state = None 
        
    def main(self): 
        fig, self.ax = plt.subplots()
        self.maxim_x = 70
        self.maxim_y = 30 
        plt.xticks(range(1,self.maxim_x))
        plt.yticks(range(1,self.maxim_y))
        plt.gca().set_aspect('equal', adjustable='box')
        self.st = state([1,3,5,7,9,10,23],[2,4,6,8,12,13,14],self.ax)
        plt.show()
        self.button_config()
    def button_config(self): 
        button = widgets.Button(description='Siguiente')
        out = widgets.Output()
        button.on_click(self.on_button_clicked)
        # displaying button and its output together
        box = widgets.VBox([button,out])
        display(box)
    def on_button_clicked(self,_):
        self.st.update() 
c = cell3()
c.main()
```
