```python
import numpy as np
import math 
import matplotlib.pyplot as plt
from matplotlib.table import table
from matplotlib.transforms import Bbox
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from matplotlib.patches import Path
from matplotlib.patches import PathPatch
import ipywidgets as widgets 
%matplotlib nbagg 

class union_find: 
    def __init__(self,n):
        self.parent = [i for i in range (0,n)]
        self.size = [1 for i in range (0,n) ]
    
    def unite(self,i,j): 
        pi = self.find(i)
        pj = self.find(j)
        if(pi == pj):
            return 
        if(self.size[pi] < self.size[pj]): 
            self.parent[pi] = pj
            self.size[pj] = self.size[pj] + self.size[pi]
        else: 
            self.parent[pj] = pi
            self.size[pi] = self.size[pi] + self.size[pj]
    
    def find(self,i): 
        while(self.parent[i] != i): 
            i = self.parent[i]
        return i
def inter_points(rad,x1,y1,x2,y2): 
    phi = math.atan2(y2-y1, x2-x1)
    x = x1 + rad * math.cos(phi)
    y = y1 + rad * math.sin(phi)
    return (x,y)

def mid_point(p1,p2): 
    (x1,y1) = p1
    (x2,y2) = p2
    return ((x1 + x2)/2.0,(y1 + y2)/2.0)

def crear_grafica(n,ars): 
    graf = [dict() for i in range(0,n)]
    for (x,y,w) in ars: 
        graf[x][y] = w
        graf[y][x] = w 
    return graf
    
def dibujar_grafica(n,graf,pos_vert,ax):
    graf_rep = [[None,dict([])] for i in range(0,n)] 
    rad = 1.0 
    #primero dibujamos los nodos
    for i in range(0,n):
        (x,y) = pos_vert[i]
        c = Circle((x,y),radius = rad,facecolor = 'white',edgecolor = 'black')
        ax.add_patch(c)
        anot = ax.annotate(str(i), (x, y),color='black', weight='bold', fontsize=7, ha='center', va='center')
        graf_rep[i][0] = (c,anot) 
    #regresa la representacion de la grafica 
    for i in range(0,n):
        for (j,w) in graf[i].items(): 
            if j not  in graf_rep[i][1]: 
                xi,yi = graf_rep[i][0][0].get_center()
                xj,yj = graf_rep[j][0][0].get_center()
                linea = PathPatch(Path([inter_points(rad,xi,yi,xj,yj),inter_points(rad,xj,yj,xi,yi)]), facecolor='none', edgecolor='black')
                ax.add_patch(linea)
                anot = ax.annotate(str(w),mid_point(graf_rep[i][0][0].get_center(),graf_rep[j][0][0].get_center()),color='black', weight='bold', fontsize=7, ha='center', va='center')
                graf_rep[i][1][j] = (linea,anot)
                graf_rep[j][1][i] = (linea,anot) 
    return graf_rep
    
def takeThird(elem): 
    return elem[2]
        
class state: 
    
    def __init__(self,ars,graf_rep,n): 
        #ordena las aristas y asignals 
        self.ars = ars
        self.ars.sort(key = takeThird)
        self.graf_rep = graf_rep
        self.union_f = union_find(n); 
        
    def update(self):  
        if(not self.ars): 
            print("acabo")
            return
        (x,y,w) = self.ars[0]
        self.ars.pop(0)
        if(self.union_f.find(x) != self.union_f.find(y)):  
            self.graf_rep[x][1][y][0].set(color = 'yellow') 
            self.graf_rep[x][1][y][0].set(linewidth = 4.0) 
            #tenemos que unir los vertices 
            self.union_f.unite(x,y)
            self.graf_rep[x][0][0].set(facecolor = 'yellow')
            self.graf_rep[y][0][0].set(facecolor = 'yellow')
        else: 
            self.graf_rep[x][1][y][0].set(visible = False)
            self.graf_rep[x][1][y][1].set(visible = False)
        
class cell: 
    def __init__(self): 
        self.ax = None
        self.maxim_x = None 
        self.maxim_y = None 
        self.st = None       
    def main(self): 
        fig, self.ax = plt.subplots()
        self.maxim_x = 20
        self.maxim_y = 20
        plt.xticks(range(1,self.maxim_x))
        plt.yticks(range(1,self.maxim_y))
        plt.gca().set_aspect('equal', adjustable='box')
        pos_vert = [(1,10),(10,10),(5.5,5),(1,1),(10,1),(6,13),(14,5),(2,13),(16,15),(17,2)]
        ars = [(0,1,1),(0,3,4),(0,2,1),(1,2,1),(1,4,1),(2,4,1),(2,3,1),(3,4,1),(5,1,2),(5,0,8),(6,4,3),(6,2,7),(7,2,2),(8,1,9),(8,4,2),(4,9,1),(6,9,2)]
        g = crear_grafica(len(pos_vert),ars)
        graf_rep = dibujar_grafica(len(pos_vert),g,pos_vert,self.ax)
        self.st = state(ars,graf_rep,len(pos_vert))
        plt.show()
        self.button_config()
    def button_config(self): 
        button = widgets.Button(description='Siguiente')
        out = widgets.Output()
        button.on_click(self.on_button_clicked)
        # displaying button and its output together
        box = widgets.VBox([button,out])
        display(box)
    def on_button_clicked(self,_):
        self.st.update() 
c = cell()
c.main()
```
