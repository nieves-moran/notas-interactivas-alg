```python
from cmath import atan
from glob import glob
from matplotlib.patches import Circle
from matplotlib.patches import Rectangle
from matplotlib.patches import Path
from matplotlib.patches import PathPatch
import matplotlib.pyplot as plt
from IPython.display import display
import ipywidgets as widgets
from IPython.display import clear_output
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import random 

import math
def inter_points(rad,x1,y1,x2,y2): 
    phi = math.atan2(y2-y1, x2-x1)
    x = x1 + rad * math.cos(phi)
    y = y1 + rad * math.sin(phi)
    return (x,y)

class Fig1:
    circ = dict()
    def config_imagen(self): 
        self.fig,self.ax = plt.subplots()
        plt.gca().set_aspect('equal', adjustable='box')
        plt.axis("off")
    def poner_circulos(self):
        self.circ = { 3 : (0,0),4 : (2,0),5 : (4,0),6 : (6,0),1 : (1,2),2 : (5,2),0 : (3,4)}
        for (id,(x,y)) in self.circ.items(): 
            c = Circle((x,y),radius = 0.5,facecolor = 'white',edgecolor = 'black')
            self.ax.add_patch(c)
        self.ax.relim()
        self.ax.autoscale_view()
    def poner_linea(self,c1,c2):
        x1,y1 = self.circ[c1]
        x2,y2 = self.circ[c2]
        p1 = inter_points(0.5,x1,y1,x2,y2) 
        p2 = inter_points(0.5,x2,y2,x1,y1)
        linea = PathPatch(Path([p1,p2]), facecolor='none', edgecolor='black')
        self.ax.add_patch(linea)
    def poner_lineas(self): 
        lineas = [(0,1),(0,2),(1,3),(1,4),(2,5),(2,6)]
        for (c1,c2) in lineas: 
            self.poner_linea(c1,c2)
    def poner_anot_circ(self): 
        anots = {0 : "$cn$", 1 : "$cn/2$", 2 : "$cn/2$" , 3 :"$cn/4$",  4 : "$cn/4$",  5 : "$cn/4$", 6 : "$cn/4$" }
        for (i,(x,y)) in self.circ.items(): 
            self.ax.text(x,y,anots[i],va = 'center',ha = 'center')
    def poner_anot_niveles(self):
        self.ax.text(8,4,"Nivel 0: $cn$",fontsize = 13)
        self.ax.text(8,2,"Nivel 1: $cn/2 + cn/2 = cn$",fontsize = 13)
        self.ax.text(8,0,"Nivel 2: $4(cn/4) = cn$",fontsize = 13)
    def __init__(self): 
        self.config_imagen() 
        self.poner_circulos()
        self.poner_lineas()
        self.poner_anot_circ() 
        self.poner_anot_niveles()
f = Fig1() 
```
